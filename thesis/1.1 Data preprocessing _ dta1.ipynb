```python
import pandas as pd
import time
import numpy as np
from tqdm import tqdm
```

```python
data=pd.read_csv('LN.txt')
```

```python
data
```

```python
data['open_time']
```

```python
TimeArray=[]
Time=[]
for i in range(len(data['open_time'])):
    timeStamp = float(data['open_time'][i])
    timeArray = time.localtime(timeStamp)
    TimeArray.append(timeArray)
    otherStyleTime = time.strftime("%Y-%m-%d %H:%M:%S", timeArray)
    Time.append(otherStyleTime)   
```

```python
TimeArray
```

```python
Time
```

```python
NodeList=[]
for i in range(len(data['open_time'])):
    NodeList.append(data['node_ID1'][i])
    NodeList.append(data['node_ID2'][i])
```

```python
NodeList
```

```python
Nodenamelist=list(set(NodeList))
Nodenamelist.sort(key=NodeList.index)
```

```python
len(Nodenamelist)
```

```python
Nodenamelist
```

```python
Nodenumber=list(range(1,31695,1))
```

```python
node_ID1_index=[]
node_ID2_index=[]
```

```python
for j in tqdm(range(len((data['node_ID1'])))):
    for i in range(len(Nodenumber)):
        if data['node_ID1'][j]==Nodenamelist[i]:
            node_ID1_index.append(Nodenumber[i])
        elif data['node_ID2'][j]==Nodenamelist[i]:
            node_ID2_index.append(Nodenumber[i])
    
```
