```python
df_open = dict(tuple(data.groupby(data['openTime'])))
```

```python
df_close= dict(tuple(data.groupby(data['closeTime'])))
```

## Step2: Make the true existing edge list for each days

Cut by days, if the edges still open, keep them in the  graph.
Remove the edges on the day it closed.

```python
make_df=pd.DataFrame()
graphdict={}
for day in df_open.keys():
    addedge=df_open[day]
    make_df=pd.concat([make_df,addedge])
    if day in df_close.keys():
        closeedge=df_close[day]
        make_df=pd.concat([make_df,closeedge,closeedge]).drop_duplicates(keep=False)
        graphdict[day]=make_df
    else:
        graphdict[day]=make_df
```

```python
graphdict
```

```python
# save dataframe dict to a file
with open("graphdicts.pkl", "wb") as fp:   #Pickling
    pickle.dump(graphdict, fp)
```

```python
# load the dict of dataframes
with open('graphdicts.pkl', 'rb') as fp:
    graphdict = pickle.load(fp)
```

## Step3: Make each dataframe into a graph and save

```python
Network_dict={}
for day in graphdict.keys():
    graph_df=graphdict[day]
    G=nx.from_pandas_edgelist(graph_df, 'node_ID1_index', 'node_ID2_index', ['weight'])
    Network_dict[day]=G
```

```python
with open("Network_dict.pkl", "wb") as fp:   #Pickling
    pickle.dump(Network_dict, fp)
```
