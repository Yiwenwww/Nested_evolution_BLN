```python
import matplotlib.pyplot as plt
import os

```

```python
# Create a 5x4 grid of plots
fig, axs = plt.subplots(5, 4, figsize=(10, 10))


# Path to the folder containing the plots
plots_folder = "C:/Users/wyw25/OneDrive/Desktop/figforpaper/simu"



# Sample y-axis labels
y_labels = ['Label 1', 'Label 2', 'Label 3', 'Label 4', 'Label 5']
x_labels = ['Label 1', 'Label 2', 'Label 3', 'Label 4']

# Plot each subplot
for i in range(5):
    for j in range(4):
        # Generate the filename for the plot
        filename = os.path.join(plots_folder, f"{i+1}{j+1}.png")
        
        # Check if the file exists
        if os.path.exists(filename):
            # Load and plot the image
            img = plt.imread(filename)
            axs[i, j].imshow(img)
            
        
# Add x and y labels to the first row and first column
for i, label in enumerate(y_labels):
    axs[i, 0].set_ylabel(label, labelpad=15, rotation=0, ha='right')

for j, label in enumerate(x_labels):
    axs[4, j].set_xlabel(label, labelpad=15)


# Adjust layout to prevent overlap of labels
plt.tight_layout()
plt.axis('off')
plt.show()
```

```python
# Create a 5x4 grid of plots
fig, axs = plt.subplots(5, 4, figsize=(10, 10))


# Path to the folder containing the plots
plots_folder = "C:/Users/wyw25/OneDrive/Desktop/figforpaper/simu"



# Sample y-axis labels
y_labels = ['0.20', '0.30', '0.40', '0.45', '0.48']
x_labels = ['0.40', '0.60', '0.80', '0.98']

# Plot each subplot
for i in range(5):
    for j in range(4):
        # Generate the filename for the plot
        filename = os.path.join(plots_folder, f"{i+1}{j+1}.png")
        
        # Check if the file exists
        if os.path.exists(filename):
            # Load and plot the image
            img = plt.imread(filename)
            axs[i, j].imshow(img)
        
        # Hide x and y axis tick labels
        axs[i, j].set_xticklabels([])
        axs[i, j].set_yticklabels([])
        axs[i, j].spines['bottom'].set_color('w')
        axs[i, j].spines['top'].set_color('w')
        axs[i, j].spines['left'].set_color('w')
        axs[i, j].spines['right'].set_color('w')
        

# Add x and y labels to the first row and first column
for i, label in enumerate(y_labels):
    axs[i, 0].set_ylabel(label, labelpad=15, rotation=0, ha='right')

for j, label in enumerate(x_labels):
    axs[4, j].set_xlabel(label, labelpad=15)

# Adjust layout to prevent overlap of labels
plt.tight_layout()

plt.savefig('simulation.pdf', format='pdf')
plt.show()
```
